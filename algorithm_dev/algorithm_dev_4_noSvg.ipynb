```python
# tile each element
# from https://stackoverflow.com/a/52346065

def broadcast_tile(a, h, w):
    x, y = a.shape
    m, n = x * h, y * w
    return np.broadcast_to(
        a.reshape(x, 1, y, 1), (x, h, y, w)
    ).reshape(m, n)

expanded = broadcast_tile(colored, 5, 5)
```
